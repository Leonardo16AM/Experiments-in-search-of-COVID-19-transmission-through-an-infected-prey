```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from pyentrp import entropy as ent

def system(x, t, a, b, c, d, R):
    dx1 = a*x[0] - R*x[0]*x[1] - a*x[0]*x[3]
    dx2 = R*x[0]*x[1] - a*x[1]*x[3] - x[1]
    dx3 = b*x[2] - a*c*x[2]*x[3]
    dx4 = d*(x[0] + x[1] + x[2])*x[3] - d*x[3]
    return [dx1, dx2, dx3, dx4]

a = 0.3
b = 0.3
c = 2
d = 0.8
k2_values = np.linspace(0.1, 0.5, 5)  # or adjust as needed
R_values = np.linspace(1.3, 1.55, 5)  # or adjust as needed
t = np.arange(0.0, 100.0, 0.01)

results = []

for k2 in k2_values:
    row = []
    for R in R_values:
        print(f"{k2} {R}")
        initial_conditions = [k2, 0.3, 0.3, 0.3]
        sol = odeint(system, initial_conditions, t, args=(a, b, c, d, R))
        sam_entropy = ent.sample_entropy(sol[:, 0], 4, 0.2 * np.std(sol[:, 0]))
        row.append(sam_entropy[3])
    results.append(row)

transposed = list(map(list, zip(*results)))
plt.figure(figsize=(10, 8))
plt.imshow(transposed, cmap='inferno', interpolation='nearest', extent=[R_values.min(), R_values.max(), k2_values.min(), k2_values.max()], origin='lower')
plt.colorbar(label='Sample Entropy')
plt.xlabel('k2')
plt.ylabel('R')
plt.title('Complexity Results of System')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pyentrp import entropy as ent
from scipy.integrate import odeint

# Funcin del sistema dinmico
def system(x, t, a, b, c, d, R):
    dx1 = a*x[0] - R*x[0]*x[1] - a*x[0]*x[3]
    dx2 = R*x[0]*x[1] - a*x[1]*x[3] - x[1]
    dx3 = b*x[2] - a*c*x[2]*x[3]
    dx4 = d*(x[0] + x[1] + x[2])*x[3] - d*x[3]
    return [dx1, dx2, dx3, dx4]

# Parmetros
a, b, c, d, R = 0.3, 0.3, 2, 0.8, 0.5

# Rango de condiciones iniciales
initial_conditions_k2 = np.linspace(0.1, 0.5, 5)
initial_conditions_R = np.linspace(1.3, 1.55, 5)

# Funcin para calcular la entropa de la muestra
def calculate_sample_entropy(initial_condition_k2, initial_condition_R):
    print(f"{initial_condition_k2} {initial_condition_R}")
    # print(initial_condition_k2+" "+ initial_condition_R)
    y0 = [initial_condition_k2, 0.3, 0.3, 0.3]  # condiciones iniciales
    t = np.arange(0.0, 100.0, 0.01)
    sol = odeint(system, y0, t, args=(a, b, c, d, initial_condition_R))
    return [ent.sample_entropy(sol[:, i], 3, 5 * np.std(sol[:, i])) for i in range(4)]





# Calcular la entropa de la muestra para cada par de condiciones iniciales
sample_entropies    = np.array([[calculate_sample_entropy(ic_k2, ic_R) for ic_R in initial_conditions_R] for ic_k2 in initial_conditions_k2])

# Crear los grficos
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
for i, ax in enumerate(axs.flatten()):
    im = ax.imshow(sample_entropies[:, :, i], cmap='plasma', origin='lower', aspect='auto', 
                   extent=[initial_conditions_R[0], initial_conditions_R[-1], 
                           initial_conditions_k2[0], initial_conditions_k2[-1]])
    ax.set_xlabel('R')
    ax.set_ylabel('k2')
    ax.set_title(f'Sample Entropy for x{i+1}')
    fig.colorbar(im, ax=ax)

plt.tight_layout()
plt.show()
```
